# 2) Drop unqualified variables

*   Drop variables with missing > 50%
*   Drop categorical variables with flat values > 70% (variables with the same value in the same column)

<font color='blue'>Q2) How many columns do we have left?</font>




```python
df.dropna(thresh=(len(df)/2),axis=1,inplace=True) #drop column that have missing value more than 50%
df_category = df[["Name","Sex","Ticket","Embarked"]] #select category variable
column_percentage = df_category.apply(lambda x: x.value_counts(normalize=True).max())
column_drop = column_percentage[column_percentage > 0.7].index #drop variable with flat values > 70%

df = df.drop(columns=column_drop)
df.shape[1]
```

# 3) Remove all rows with missing target (the variable "Survived")

<font color='blue'>Q3) How many rows do we have left?</font>

```python
df.dropna(subset="Survived",inplace=True) #remove row that column survived is null
df.shape[0] #display number of row without missing target
```

# 4) Handle outliers

For the variable “Fare”, replace outlier values with the boundary values


*   If value < (Q1 - 1.5IQR), relace with (Q1 - 1.5IQR)
*   If value > (Q3 + 1.5IQR), relace with (Q3 + 1.5IQR)

<font color='blue'>Q4) What is the mean of “Fare” after replacing the outliers (round 2 decimal points)?</font>



```python
#calculate Q1,Q3 and IQR for column Fare
quantile_1 = df["Fare"].quantile(0.25)
quantile_3 = df["Fare"].quantile(0.75)
iqr = quantile_3 - quantile_1

#calculate lower_bound and upper_bound for outlier replacement
lower_bound = quantile_1 - (1.5 * iqr)
upper_bound = quantile_3 + (1.5 * iqr)

df.loc[df.Fare < lower_bound , "Fare"] = lower_bound #replaced outlier by lower_bound
df.loc[df.Fare > upper_bound , "Fare"] = upper_bound #replaced outlier by upper_bound
df #show table
```

```python
df["Fare"].mean().round(2) #display mean of column Fare
```

# 5) Impute missing value



*   Show the number of missing values in each variable
*   Impute missing values with mean & mode
*   Show the number of missing values again after missing value imputation

<font color='blue'>Q5) Which variable has the largest number of missing values?</font>



```python
df.isnull().sum() #show number of missing value each column
```

```python
df[["Pclass","Age","SibSp"]] = df[["Pclass","Age","SibSp"]].fillna(df.median())
df = df.fillna(df.mean())
```

```python
df.isnull().sum() #show number of missing value after imputation
```

# 6) Convert categorical to numeric values



*   Drop the variables “Name” & “Ticket”
*   For the variables “Sex” & “Embarked”, perform the dummy coding and drop the first level. Also, drop those original variables (“Sex” & “Embarked”)

<font color='blue'>Q6) How many columns do we have?</font>



```python
df = df.drop(["Name","Ticket"],axis=1) #drop name and ticket column
nominal_columns = ["Sex"] #set Sex to nominal category
dummy_df = pd.get_dummies(df[nominal_columns], drop_first=True) #set Sex column to dummy_code
df_dummy = pd.concat([df, dummy_df], axis=1)
df_dummy = df_dummy.drop(nominal_columns, axis=1)
df_dummy #show table
```
