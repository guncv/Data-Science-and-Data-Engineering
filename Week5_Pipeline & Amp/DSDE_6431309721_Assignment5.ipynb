```
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

# 1. Load the "ModifiedEdibleMushroom.csv" data

```
# Load data
df = pd.read_csv('https://raw.githubusercontent.com/pvateekul/2110446_DSDE_2023s2/main/DataSet/hed2020_dataset.csv')
```

```
df.head() #show top 5 row of data
```

# 2. Choose edible mushrooms only.

```
df_edible = df[df.label == 'e'] #choose only edible mushroom
len(df_edible) #number of row
```

```
df_edible.head() #show top 5 row of data
```

#3.  Only the variables below have been selected to describe the distinctive characteristics of edible mushrooms
  - 'cap-color-rate','stalk-color-above-ring-rate'


```
df_cap = df_edible[["cap-color-rate","stalk-color-above-ring-rate"]] #pick only 2 column
```

```
df_cap.head() #show example
```

#4. Provide a proper data preprocessing as follows:
  - Fill missing with mean
  - Standardize variables with Standard Scaler

```
df_cap.isnull().sum() #number of row that is null
```

```
df_cap.fillna(df_cap.mean(), inplace=True) #fill null with mean value
```

```
df_cap.isnull().sum() #check number of row that is null again
```

```
scaler = StandardScaler()
df_cap[["cap-color-rate","stalk-color-above-ring-rate"]] = scaler.fit_transform(df_cap[:])
```

#5. K-means clustering with 5 clusters and seed = 2020.


```
kmeans = KMeans(n_clusters=5, random_state=2020).fit(df_cap) # initializes and fits a KMeans clustering
```

```
kmeans # show kmean
```

```
df_cap["cluster"] = kmeans.labels_ #add column cluster
```

```
df_cap.head() # show top 5 row of df_cap
```

#6. Show the centroid of each cluster.

```
# creates a DataFrame named "centroids" containing the coordinates of the cluster
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=["cap-color-rate","stalk-color-above-ring-rate"])
```

```
centroids # show centroid
```

# 7. Convert the centroid value to the original scale

```
# convert to original scale
centroids[["cap-color-rate","stalk-color-above-ring-rate"]] = scaler.inverse_transform(centroids[["cap-color-rate","stalk-color-above-ring-rate"]])
```

```
centroids # show convered centroid
```
